```python

myString = "1 1 2 1.01 3 1.05 4 1.14 5 1.24 6 1.34 7 1.43 8 1.50 9 1.56 10 1.61 15 1.78 20 1.87 30 1.98 40 2.06 50 2.14 60 2.21 90 2.36"



mySplitString = myString.split(" ")



i = 0
greenData = dict()

while i < len(mySplitString) - 1:
    distance = float(mySplitString[i])
    i += 1
    value = float(mySplitString[i])
    i += 1

    greenData[distance] = value










import matplotlib.pyplot as plt

green = list()
greenYardages = list()

for yardageVal in greenData.keys():
    green.append(greenData[yardageVal])
    greenYardages.append(yardageVal)



plt.scatter(greenYardages, green, label="Green")


plt.xlabel('Feet from Hole')

plt.ylabel('Expected Shots to Hole out')

# giving a title to my graph
plt.title('Putting graph')

plt.legend()

# function to show the plot
plt.show()

import numpy as np
import matplotlib.pyplot as plt

# Convert the lists to numpy arrays
x = np.array(greenYardages)
y_green = np.array(green)

# Fit a polynomial of order 5
coeffs_green = np.polyfit(x, y_green, 5)


# Print the coefficients
print("Green Coefficients:", coeffs_green)


# Generate the x values for the line of best fit
x_fit = np.linspace(min(x), max(x), 100)

# Evaluate the polynomial at x_fit
y_fit_green = np.polyval(coeffs_green, x_fit)


# Scatter plot of the data
plt.scatter(x, y_green, label="Green")
# plt.scatter(x, y_rough, label="Rough")
# plt.scatter(x, y_sand, label="Sand")
# plt.scatter(x, y_recovery, label="Recovery")
# plt.scatter(teeYardages, y_tee, label="Tee")

# Line of best fit
plt.plot(x_fit, y_fit_green, 'r', label="Green Fit")
# plt.plot(x_fit, y_fit_rough, 'g', label="Rough Fit")
# plt.plot(x_fit, y_fit_sand, 'b', label="Sand Fit")
# plt.plot(x_fit, y_fit_recovery, 'm', label="Recovery Fit")
# plt.plot(x_fit, y_fit_tee, 'c', label="Tee Fit")

plt.xlabel('Feet from Hole')
plt.ylabel('Expected Shots to Hole out')
plt.title('Green shot graph')
plt.legend()
plt.show()
```
