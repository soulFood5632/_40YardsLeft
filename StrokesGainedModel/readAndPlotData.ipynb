In the below cell we will convert this list of string of data into a list of float point values

```python
valueList1 = list()
valueList2 = list()

for strVal in mySplitString:
    valueList1.append(float(strVal))


for strVal in otherSplitString:
    valueList2.append(float(strVal))


```

In the next cell we will collect the data to become a training database

```python
teeData = dict()
fairwayData = dict()
roughData = dict()
sandData = dict()
recoveryData = dict()

string1 = 0
string2 = 0
while string2 < otherSplitString.__len__():
    yardage = valueList1[string1]
    string1 += 1
    if yardage >= 100:
        teeData[yardage] = valueList1[string1]
        string1 += 1

    fairwayData[yardage] = valueList1[string1]
    string1 += 1

    roughData[yardage] = valueList2[string2]
    string2 += 1

    sandData[yardage] = valueList2[string2]
    string2 += 1

    recoveryData[yardage] = valueList2[string2]
    string2 += 1

```

In the below cell we will plot this data

```python
import matplotlib.pyplot as plt

teeYardages = list()
yardages = list()
tee = list()
fairway = list()
rough = list()
sand = list()
recovery = list()

for yardageVal in teeData.keys():
    tee.append(teeData[yardageVal])
    teeYardages.append(yardageVal)

for yardageVal in fairwayData.keys():
    yardages.append(yardageVal)
    fairway.append(fairwayData[yardageVal])
    rough.append(roughData[yardageVal])
    sand.append(sandData[yardageVal])
    recovery.append(recoveryData[yardageVal])



plt.scatter(yardages, fairway, label = "Fairway")
plt.scatter(yardages, rough, label = "Rough")
plt.scatter(yardages, sand, label = "Sand")
plt.scatter(yardages, recovery, label = "Recovery")
plt.scatter(teeYardages, tee, label = "Tee")

plt.xlabel('Yardage from Hole')

plt.ylabel('Expected Shots to Hole out')

# giving a title to my graph
plt.title('Tee shot graph')

plt.legend()

# function to show the plot
plt.show()

```

In the below cell we contruct a polynomial regression model

```python
import numpy as np
import matplotlib.pyplot as plt

# Convert the lists to numpy arrays
x = np.array(yardages)
y_fairway = np.array(fairway)
y_rough = np.array(rough)
y_sand = np.array(sand)
y_recovery = np.array(recovery)
y_tee = np.array(tee)

# Fit a polynomial of order 5
coeffs_fairway = np.polyfit(x, y_fairway, 9)
coeffs_rough = np.polyfit(x, y_rough, 8)
coeffs_sand = np.polyfit(x, y_sand, 10)
coeffs_recovery = np.polyfit(x, y_recovery, 12)
coeffs_tee = np.polyfit(teeYardages, y_tee, 9)


# Print the coefficients
print("Fairway Coefficients:", coeffs_fairway)
print("Rough Coefficients:", coeffs_rough)
print("Sand Coefficients:", coeffs_sand)
print("Recovery Coefficients:", coeffs_recovery)
print("Tee Coefficients:", coeffs_tee)

# Generate the x values for the line of best fit
x_fit = np.linspace(min(x), max(x), 100)

# Evaluate the polynomial at x_fit
y_fit_fairway = np.polyval(coeffs_fairway, x_fit)
y_fit_rough = np.polyval(coeffs_rough, x_fit)
y_fit_sand = np.polyval(coeffs_sand, x_fit)
y_fit_recovery = np.polyval(coeffs_recovery, x_fit)
y_fit_tee = np.polyval(coeffs_tee, x_fit)

# Scatter plot of the data
# plt.scatter(x, y_fairway, label="Fairway")
# plt.scatter(x, y_rough, label="Rough")
# plt.scatter(x, y_sand, label="Sand")
plt.scatter(x, y_recovery, label="Recovery")
# plt.scatter(teeYardages, y_tee, label="Tee")

# Line of best fit
# plt.plot(x_fit, y_fit_fairway, 'r', label="Fairway Fit")
# plt.plot(x_fit, y_fit_rough, 'g', label="Rough Fit")
# plt.plot(x_fit, y_fit_sand, 'b', label="Sand Fit")
plt.plot(x_fit, y_fit_recovery, 'm', label="Recovery Fit")
# plt.plot(x_fit, y_fit_tee, 'c', label="Tee Fit")

plt.xlabel('Yardage from Hole')
plt.ylabel('Expected Shots to Hole out')
plt.title('Tee shot graph')
plt.legend()
plt.show()
```
